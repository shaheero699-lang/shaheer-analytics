```python
# Revenue Forecasting Model

## Veltrix RevOps Intelligence

This notebook demonstrates revenue forecasting and trend analysis using historical sales data.

### Objectives
- Analyze revenue trends
- Forecast future performance
- Support strategic planning
- Identify growth opportunities
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Forecasting Notebook Loaded Successfully")
```

```python
revenue = pd.DataFrame({
    'Month':['Jan','Feb','Mar','Apr','May','Jun'],
    'Revenue':[12000,14000,13500,16000,18000,19500]
})

revenue
```

```python
plt.figure(figsize=(8,4))
plt.plot(revenue['Month'], revenue['Revenue'])
plt.title("Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
avg_growth = revenue['Revenue'].pct_change().mean()

print("Average Monthly Growth Rate:", round(avg_growth*100,2), "%")
```
