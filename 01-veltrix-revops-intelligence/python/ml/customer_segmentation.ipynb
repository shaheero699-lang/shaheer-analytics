```python
# Customer Segmentation

## Veltrix RevOps Intelligence

This notebook segments customers based on customer health scores.

### Objectives
- Identify high-risk customers
- Classify customer health
- Support retention initiatives
- Improve customer success operations
```

```python
import pandas as pd
import numpy as np

print("Customer Segmentation Notebook Loaded Successfully")
```

```python
customers = pd.DataFrame({
    'customer_id':[101,102,103,104,105,106],
    'customer_health_score':[95,82,67,48,73,35]
})

customers
```

```python
customers['health_segment'] = pd.cut(
    customers['customer_health_score'],
    bins=[0,50,75,100],
    labels=['At Risk','Moderate','Healthy']
)

customers
```

```python
customers['health_segment'].value_counts()
```

```python
segment_summary = customers.groupby(
    'health_segment',
    observed=False
).agg({
    'customer_id':'count',
    'customer_health_score':'mean'
})

segment_summary
```
